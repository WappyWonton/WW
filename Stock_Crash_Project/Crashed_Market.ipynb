```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as mtick
#import seaborn as sns
```

```python
df = pd.read_csv('Crashed_Values_Perecent.csv') 
```

```python
plt.plot('Days', '1929', data=df, color='grey')
plt.plot('Days', '1987', data=df, color='black')
plt.plot('Days', '2008', data=df, color='orange')
plt.plot('Days', '2020', data=df, color='red')
plt.legend()

plt.xlabel('Days From Peak')
plt.ylabel('%')
plt.title('S&P 500, 500 days after peaks in historical crises')
plt.text(100, -62, 'Source: Yahoo Finance Historical Data', style='italic', fontsize=10)
plt.savefig('filename.png', dpi=3000)
```
